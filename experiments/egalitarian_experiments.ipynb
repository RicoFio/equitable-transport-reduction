```python
import numpy as np

import matplotlib.pyplot as plt
import igraph as ig
from eptnr.datasets.dataset_loader import load_dataset
from eptnr.datasets.synthetic_datasets import SyntheticDatasets
from eptnr.rewards.egalitarian import EgalitarianTheilAndCostReward
from eptnr.constants.travel_metric import TravelMetric
from eptnr.rewards.utils.graph_computation_utils import get_tt_hops_com_dfs
from eptnr.analysis.full_problem_exploration import full_problem_exploration
import pandas as pd
from eptnr.algorithms.baselines.optimal_baseline import optimal_max_baseline
import seaborn as sns
```

# Dataset loading

```python
graph, census = load_dataset(SyntheticDatasets.ONE)
```

# Data Anaysis

```python
def plot_travel_time_histogram(graph, census):
    tt_df, _, _ = get_tt_hops_com_dfs(graph, census, 50)
    temp_tt_df = tt_df.copy()
    categories = pd.Categorical(temp_tt_df['group'])
    temp_tt_df['group'] = categories
    temp_tt_df = temp_tt_df.pivot(columns='group')

    melted_temp_tt = temp_tt_df.melt()[['group', 'value']]
    melted_temp_tt.value = pd.to_numeric(melted_temp_tt.value)
    ax = sns.histplot(melted_temp_tt, x='value', hue='group', multiple='dodge', shrink=.75, bins=20,
                      palette=categories.unique().to_list())
    ax.vlines(melted_temp_tt.value.mean(), ymin=0, ymax=melted_temp_tt.groupby('group').count().max(), linestyles="dashed")
    return ax
```

```python
plot_travel_time_histogram(graph, census)
```

# Reward setup

```python
graph.es.select(type_ne='walk')['cost']
```

```python
total_graph_cost = sum(graph.es.select(type_ne='walk')['cost'])
total_graph_cost
```

```python
max_travel_time = max(graph.es.select(type_ne='walk')['tt'])
```

```python
monetary_budget = 200
groups = None
metrics = [TravelMetric.TT]


reward = EgalitarianTheilAndCostReward(census_data=census, com_threshold=0,
                                       total_graph_cost=total_graph_cost, monetary_budget=monetary_budget,
                                       groups=groups, metrics=metrics, verbose=True)
```

# Algorithms

```python
edge_types = set(graph.es.select(type_ne='walk')['type'])
max_solutions = optimal_max_baseline(graph, reward, edge_types=edge_types)
```

```python
max_solutions
```

# Evaluation

```python
graph_prime = graph.subgraph_edges([e.index for e in graph.es if e.index not in max_solutions[0][1]])
ig.plot(graph_prime)
```

```python
plot_travel_time_histogram(graph_prime, census)
```

# Max Q

```python
from eptnr.algorithms.deep_rl.deep_q_learning.deep_max_q_learner import DeepMaxQLearner
```

```python
episodes = 128
batch_size = 64
replay_memory_size = 2048
target_network_update_step = 16
static_eps_steps = 64

dmql = DeepMaxQLearner(
    base_graph=graph,
    reward=reward,
    edge_types=edge_types,
    episodes=episodes,
    batch_size=batch_size,
    replay_memory_size=replay_memory_size,
    target_network_update_step=target_network_update_step,
    static_eps_steps=static_eps_steps
)
```

```python
max_rewards_over_episodes, epsilon_over_episodes = dmql.train(return_max_rewards_over_episodes=True,return_epsilon_over_episodes=True,verbose=False)
```

```python
plt.plot(max_rewards_over_episodes)
plt.plot(epsilon_over_episodes)
plt.plot(dmql.policy_net_loss[::4])
```

```python
dmql_solutions = dmql.inference()
```

```python
graph_prime = graph.subgraph_edges([e.index for e in graph.es if e.index not in dmql_solutions[1]])
ig.plot(graph_prime)
```
